```python
#plotting multiple Lines
```

```python
years = range(2000,2009) #2012 not included
apples = [0.895, 0.91, 0.919, 0.926, 0.929, 0.931, 0.934, 0.936, 0.938]
oranges = [0.962, 0.964, 0.930, 0.923, 0.919, 0.908, 0.907, 0.904, 0.901]
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)
plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)');
#all the shape must be same => apples, years, oranges have to be same shape
```

```python
#this graph is not to much clear, which line represent what so we want to be more clear
#Chart Title and Legend
```

```python
plt.plot(years, apples)
plt.plot(years, oranges)

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title('Crop yields in somewhere')
plt.legend(['Apples', 'Oranges']);
#plt.legend sayesinde hangi istatistiğin kime ait olduğunu anlayabiliyoruz.
```

```python
#LINE MARKER
plt.plot(years, apples, marker= 'o')
plt.plot(years, oranges, marker= 'v')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title('Crop yields in somewhere')
plt.legend(['Apples', 'Oranges']);
```

```python
plt.plot(years, apples, marker= 'o', color = 'lightpink', linestyle = '-')
plt.plot(years, oranges, marker= 'v', color = 'blueviolet', linestyle = '--')

plt.xlabel('Year')
plt.ylabel('Yield (tons per hectare)')

plt.title('Crop yields in somewhere')
plt.legend(['Apples', 'Oranges']);
```
